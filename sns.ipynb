```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mpg_df = sns.load_dataset("mpg")
mpg_df.head(10)
```

```python
ax = sns.scatterplot(x="weight",y="mpg",data=mpg_df)
sns.set(style='ticks')
```

```python
sns.jointplot(mpg_df.weight,mpg_df.mpg,kind = "hex",color='m')
sns.set_style("white")
```

```python
df1 = sns.load_dataset('flights')
df1.head(10)
```

```python
df_pivoted = df1.pivot("month", "year", "passengers")
df_pivoted
```

```python
ax = sns.heatmap(df_pivoted)
```

```python
sns.boxplot(x='year',y='passengers',data=df1)
```

```python
mpg_df['model_decade'] = np.floor(mpg_df.model_year/10)*10
mpg_df['model_decade'] = mpg_df['model_decade'].astype(int)
sns.violinplot(x='model_decade', y='mpg', data=mpg_df,hue='origin')
```
